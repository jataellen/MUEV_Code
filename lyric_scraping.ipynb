# Scraping Lyrics by Artist

### Access via Lyrics Genius
Accessible here: https://lyricsgenius.readthedocs.io/en/master/

```python
import lyricsgenius as lg
import pandas as pd
```

```python
api_key = "NW4oeixy7UQTEz2qsT8eJUvWtvoXhF8Dd5nUgIe13-iCUbARwRL1MS8OjQzDLdUY"
genius = lg.Genius(api_key, skip_non_songs=True, excluded_terms=["(Remix)", "(Live)"], remove_section_headers=True)
```

#### Specify artist

```python
artist = "Dead Obies"
```

```python
all_songs_by_artist = genius.search_artist(artist)
```

### Create df of songs

```python
songs = all_songs_by_artist.songs

data = [{'title': song.title, 'artist': song.artist, 'lyrics': song.lyrics} for song in songs]
df = pd.DataFrame(data)

lyrics_data = []

for index, row in df.iterrows():
    lyrics = row['lyrics']
    for line in lyrics.split("\n"):
        if line.strip() is not None and line.strip() != "":
            lyrics_data.append({
                'title': row['title'],
                'artist': row['artist'],
                'lyric': line,
                'verb_codeswitch': "",
                'uninflected': "",
                'codeswitched_verb_uninflected': "",
                'codeswitched_verb_inflected': "",
                'note': ""
            })

lyrics_df = pd.DataFrame(lyrics_data)
```

```python
lyrics_df = lyrics_df.drop_duplicates().dropna()
```

```python
lyrics_df.to_csv(f"data/rap_lyrics_{artist}.csv", index=False)
```
